## Background and Motivation ##

The Great Salt Lake has experienced an extended reduction in water levels over the past few decades, reaching a historic low in recent years. The impacts of a shrinking lake are far-reaching. As the Great Salt Lake’s water levels decrease, salinity rises, disrupting the ecosystems and the industries tied to the lake. Most concerning is the possibility of toxic dust and sediments blanketing the surrounding region and communities throughout the state as more lakebed area is slowly exposed. With these concerns in mind, recent public conversations have involved “praying for rain” and hopeful wishes for a healthier lake and wetter climate as part of a tangible solution. Moreover, these discussions raise questions around the role that increased natural precipitation might play in restoring the lake to healthier conditions.  

With this project, we hope to answer the question, how plausible is a natural recovery to historically normal water levels driven by precipitation alone? In other words, if we simply hope for more rain and snow while maintaining all other human activities impacting the lake, how likely is natural precipitation alone to restore the Great Salt Lake to target elevations within X years? We plan to determine long-term trends using USGS monitoring station data and historical weather records to estimate the probability that natural precipitation could return the lake to healthy thresholds. By evaluating improvement scenarios for the Great Salt Lake statistically instead of rhetorically, we can address a very visible and concerning environmental issue with numerical evidence.

## Ethical Considerations ##

The sharp decline in the health of the Great Salt Lake has implications and ethical considerations for a wide range of stakeholders. First and foremost, the citizens of northern Utah, primarily those living along the Wasatch Front, have the potential to be directly affected by the exposed lakebed. It is predicted that decreasing lake levels will lead to an increase in air pollution due to airborne dust and toxins from the lake. If recent trends continue, increased exposure to the unhealthy lake sediments could affect the physical health, financial interests, and quality of life of Utah residents. In addition, indigenous groups throughout the state have cultural and historical ties the Great Salt Lake and its tributaries that far predate colonial activities. These people have an ethical stake in all decisions potentially affecting the future of the lake and the ecosystems it supports. 

The Great Salt Lake is at the center of many important industries in Utah, making it a local economic driver, and an important source of water in an arid climate. This raises important questions about water resource allocation, while keeping environmental preservation, and economic activities in mind. Perhaps most consequentially, the state government and policymakers have the job of balancing water demands between urban development, agriculture, and industry to ensure the long-term health of the lake. Since this project uses real data for modeling future situations, additional ethical considerations include responsible communication and transparency around analysis results, assumptions, and models to avoid misinforming interested parties. Finally, this project does not seek to criticize religious practices, spiritual beliefs, or traditions, but instead attempts to highlight the scale of practical actions that must be taken to save the Great Salt Lake.
