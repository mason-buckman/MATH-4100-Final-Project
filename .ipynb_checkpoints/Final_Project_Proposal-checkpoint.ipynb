## Background and Motivation ##

The Great Salt Lake has experienced a prolonged decline in water levels over the past few decades, reaching a historic low in recent years. The impacts of a shrinking lake are far-reaching. As the Great Salt Lake’s water levels decrease, salinity rises, disrupting the ecosystems and the industries tied to the lake. Most concerning is the possibility of toxic dust and sediments blanketing the surrounding region and communities throughout the state as more lakebed area is slowly exposed. With these concerns in mind, recent public conversations have involved “praying for rain” and hopeful wishes for a healthier lake and wetter climate as part of a tangible solution. Moreover, these discussions raise questions around the role that increased natural precipitation might play in restoring the lake to healthier conditions.  

With this project, we aim to answer the question: How plausible is a natural recovery to historically normal water levels driven by precipitation alone? In other words, if we simply hope for more rain and snow while maintaining all other human activities impacting the lake, how likely is natural precipitation alone to restore the Great Salt Lake to target elevations within X years? We plan to determine long-term trends using USGS monitoring station data and historical weather records to estimate the probability that natural precipitation could return the lake to healthy thresholds. By evaluating improvement scenarios for the Great Salt Lake statistically rather than rhetorically, we can address a very visible and concerning environmental issue with numerical evidence.

## Ethical Considerations ##

The sharp decline in the health of the Great Salt Lake has implications and ethical considerations for a wide range of stakeholders. First and foremost, the citizens of northern Utah, primarily those living along the Wasatch Front, have the potential to be directly affected by the exposed lakebed. It is predicted that declining lake levels will increase air pollution from airborne dust and toxins released by the lake. If recent trends continue, increased exposure to the unhealthy lake sediments could affect the physical health, financial interests, and quality of life of Utah residents. In addition, indigenous groups throughout the state have cultural and historical ties to the Great Salt Lake and its tributaries that far predate colonial activities. These people have an ethical stake in all decisions that could affect the future of the lake and the ecosystems it supports. 

The Great Salt Lake is at the center of many important industries in Utah, making it a local economic driver and an important source of water in an arid climate. This raises important questions about water resource allocation, while keeping environmental preservation and economic activities in mind. Perhaps most importantly, the state government and policymakers have the job ofust balance water demands between urban development, agriculture, and industry to ensure the long-term health of the lake. Since this project uses real data to model future situations, additional ethical considerations include responsible communication and transparency around analysis results, assumptions, and models to avoid misinforming interested parties. Finally, this project does not seek to criticize religious practices, spiritual beliefs, or traditions. Instead, it attempts to highlight the scale of practical actions that must be taken to save the Great Salt Lake.

We will be extracting data from multiple sources:<br>

1. USGS.gov API to obtain historical lake elevation data
    - Saltair Boat Harbor monitoring station
2. USGS.gov API to obtain historical snowpack data
3. NOAA API using an API key to obtain daily rainfall data for several decades 
    - Salt Lake City Airport monitoring station 

Data processing will be implemented by first gathering all the raw data from the listed sources. Since we plan to gather data via API calls, we will obtain the datasets once, convert them from .json format to Pandas data frames, and then save the data frames as .csv files. Doing this will allow us to easily access our data throughout the project without having to request information from the NOAA and USGS APIs repeatedly.

After obtaining and saving our raw data as .csv files, we plan to explore and clean the data sets. Each .csv will be reimported as a Pandas data frame and checked for missing values. We plan to drop missing values as long as they do not constitue a large portion of the dataset.  
Once we have removed missing information, we will obtain descriptive statistics for the quantitative data columns. Doing so will allow us to see potential correlations between the variables we will be analyzing. 

If the data sets contain categorical variables, we will create tracking columns with numerical values so that they can be included in the analysis (Ex: Spring, Summer, Fall, Winter become 1, 2, 3, 4). With the NOAA data, we will also be able to create a new data frame containing calculated variables such as monthly cumulative and monthly average rainfall totals which will be useful for the analysis. Finally, the cleaned data frames will be re-saved as .csv files so that it can be easily accessed. If data across the separate data frames can be combined effectively into a single data frame or few larger ones, we will consider doing so to stay organized.

## Analysis Methodology ##

To evaluate whether natural precipitation alone could realistically and meaningfully improve the Great Salt Lake, we will analyze historical lake elevation and precipitation data obtained from USGS monitoring stations and climate records. Our analysis will focus on identifying relationships between precipitation patterns and changes in lake elevation over time. We plan to use a combination of exploratory data analysis, regression modeling, and simulations as follows to discover the following:

1. Is there a correlation between any of the variables (e.g., snowpack, etc.) that Utah currently tracks?  
    - Perform correlation analysis to identify variables that are relevant to drive meaningful policies and identify predictor variables.
2) Based on current precipitation and water consumption trends, when will the Great Salt Lake run dry?
    - Hypothesis Testing: Confirm or deny that the lake level decline is statistically significant or mostly natural variation.
    - Linear Regression: Create a linear regression model using Great Salt Lake elevation as the outcome and variables such as annual precipitation, snowpack levels, and water consumption as predictors to understand how much each variable contributes to the level change.  
3) Compared to a typical year, how much more precipitation would Utah need over the next 10, 15, or 20 years to replenish the Great Salt Lake?
    - Normal Distribution: Calculate how much precipitation is needed each year to reach a healthy lake level within 10, 15, or 20+ year windows and find the z-score of each scenario to determine how likely it is for that to occur in each time frame.
    - Monte Carlo Analysis (if possible): Fit a regression between precipitation and lake level change, then simulate hundreds or thousands of future precipitation events.
        - Simulated lake level trajectories can be visualized, and the probability of the lake reaching a specific level could be determined.
    - Logistic Regression (if needed): Perform a logistic regression after the linear regression to determine if this type of model forms a better fit with the data.   
4) Can data modeling be utilized to guide policy decision-making regarding water use and conservation to save the Great Salt Lake?  
    - Classification/decision tree: Set thresholds based on predictor variables (from the regression(s)) to create a decision tree that determines lake outcomes and shows how different policies could influence them. 
5) Based on precipitation and snowpack, how can historical precipitation data be categorized? 
    - Clustering: Group historical precipitation into categories (e.g., drought, average precipitation, flooding, etc.)
